```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy.random as npr
from scipy.stats import norm

def spl(N=100):
    samp = npr.exponential(size=N)
    return samp.mean()

def more_spl(N=100, sz=300):
    val_arr = np.zeros(sz)
    for i in range(sz):
        x = spl(N)  
        val = np.sqrt(N) * (x - 1)
        val_arr[i] = val
    plt.hist(val_arr, bins=30, range=(-3,3), label='experimental')
    plt.title("Histogram for N={}".format(N))
    x = np.linspace(-3,3)
    y = norm.pdf(x, 0, 1) * sz * .2
    plt.plot(x,y, label='theoretical')
    plt.legend()
    sm.qqplot(val_arr, line="45")
    plt.title("QQ plot for N={}".format(N))
```

```python
more_spl(100)
```

```python
more_spl(300)
```

```python
more_spl(1000)
```

```python
def estimate_Q(sz=1000000, N=1000):
    lmd = 1
    x = npr.gamma(N, lmd, size=sz) / N
    q = np.sum(x - 1/lmd > 0.1)
    decay_q = 1/N * np.log(q/sz)
    return q, decay_q
```

```python
print(estimate_Q(sz=100000000, N=2500))
```

```python
print(estimate_Q())
```
